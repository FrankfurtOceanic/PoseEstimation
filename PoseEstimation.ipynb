## 1. Making detections

```python
cap = cv2.VideoCapture(0)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
  while cap.isOpened():
    ret, frame = cap.read()

    
    # Convert the BGR image to RGB and process it with MediaPipe Pose.
    results = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    #instanciate what will be the annotated image
    annotated_image = frame.copy()
    #Render the image with annotations
    mp_drawing.draw_landmarks(annotated_image,
                              results.pose_landmarks, 
                              mp_pose.POSE_CONNECTIONS,
                              landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style(),
                              
                            )


    cv2.imshow("Video Capture Feed", annotated_image)

    if cv2.waitKey(5) & 0xFF == ord('q'):
      break
  cap.release()
  cv2.destroyAllWindows()
```

## 2. Determining Joints

```python
cap = cv2.VideoCapture(0)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
  while cap.isOpened():
    ret, frame = cap.read()

    
    # Convert the BGR image to RGB and process it with MediaPipe Pose.
    results = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    #instanciate what will be the annotated image
    annotated_image = frame.copy()
    
    
    # Extracting each landmark 
    try:    #the use of a try block is essential since sometimes a pose can't be detected
        
        #each landmark is stored as an array of 3 values (x,y,z) which represent the position in
        #real-world 3D coordinates in meters with the origin at the center between hips.
        landmarks = results.pose_landmarks.landmark
        
        
    except:
        pass
    
    #Render the image with annotations
    mp_drawing.draw_landmarks(annotated_image,
                              results.pose_landmarks, 
                              mp_pose.POSE_CONNECTIONS,
                              landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style(),
                            )


    cv2.imshow("Video Capture Feed", annotated_image)

    if cv2.waitKey(5) & 0xFF == ord('q'):
      break
  cap.release()
  cv2.destroyAllWindows()
```

## 3. Determing Angle

```python
def calculateAngle(a, b, c): #Simple function that calculates the angle between  point A ->B ->C
    
    
    
    g = a[:2]-b[:2] # getting the vector from b to a
    f = c[:2]-b[:2] # getting the vector from b to c

    rads = np.arccos((np.dot(g,f) / (np.linalg.norm(g)*np.linalg.norm(f))))
    angle = rads*180/np.pi
    return angle
```

### Visualizer

```python
cap = cv2.VideoCapture(0)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
  while cap.isOpened():
    ret, frame = cap.read()

    
    # Convert the BGR image to RGB and process it with MediaPipe Pose.
    results = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    #instanciate what will be the annotated image
    annotated_image = frame.copy()
    
    
    # Extracting each landmark 
    try:    #the use of a try block is essential since sometimes a pose can't be detected
        
        #each landmark is stored as an array of 3 values (x,y,z) which represent the position in
        #real-world 3D coordinates in meters with the origin at the center between hips.
        landmarks = results.pose_landmarks.landmark
        
        #get coordiantes for teh left shoulder, elbow, wrist
        LShoulder = np.array([landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].z])
        LElbow = np.array([landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].z])
        LWrist = np.array([landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].z])
        
        #now we can calculate the angle and add it to out visualization
        angle = calculateAngle(LShoulder, LElbow, LWrist)
        imgDim = annotated_image.shape
        print(tuple(np.multiply(LElbow[:2] , imgDim[:2]).astype(int)))
        annotated_image = cv2.putText(annotated_image, str(angle),
                    tuple(np.multiply(LElbow[:2] , imgDim[:2]).astype(int)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2, cv2.LINE_AA
                   )
        
    except:
        pass
    
    #Render the image with annotations
    mp_drawing.draw_landmarks(annotated_image,
                              results.pose_landmarks, 
                              mp_pose.POSE_CONNECTIONS,
                              landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style()
                            )


    cv2.imshow("Video Capture Feed", annotated_image)

    if cv2.waitKey(5) & 0xFF == ord('q'):
      break
  cap.release()
  cv2.destroyAllWindows()
```
